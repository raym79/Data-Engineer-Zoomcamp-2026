## Handling Data Types
**current issue:** data type is mixed.  
For example, `VendorID` suppose to be integer but is now float64. `tpep_pickup_datetime` suppose to be datetime but is object. 

```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates
)
```

## Ingesting Data into Postgres
**Next: we want to put them into our Postgres**  
In order to do so, we need to use `SQLAlchemy` package

**Install SQLAlchemy**

```python
# !uv add sqlalchemy
```

Running above command on jupyter notebook is as sames as running in terminal.  
we can double check in pyproject.toml file. `sqlalchemy` is now one of dependencies.

**Create Database Connection**

```python
# !uv add psycopg2
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

**Get DDL Schema**

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

**Create the Table**

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

head(n=0) makes sure we only create the table, we don't add any data yet.

At this point, the schema and table are craeted.  
To check, go to terminal run `uv run pgcli -h localhost -p 5432 -u root -d ny_taxi`

## Ingesting Data in Chunks
We don't want to insert all the data at once. Let's do it in batches and use an iterator for that:

```python
# Ingesting Data in Chunks
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)
```

**Iterate Over Chunks**

```python
# df = next(df_iter)
# df
```

```python
for df_chunk in df_iter:
    print(len(df_chunk))
```

**Insert data** 

```python
# one chunk
# df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```

```python
# !uv add tqdm
```

```python
from tqdm.auto import tqdm
```

```python
# loop
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```

okay. so now we inset all data from csv file to target table.
